```python
df = df.drop('quality', axis=1)
df.columns
```

```python
value_mapping = {'asphalt_road': 2, 'dirt_road': 0, 'cobblestone_road': 1}

# Use the map() function to apply the mapping to the column
df['y_pred'] = df['y_pred'].map(value_mapping)
df['road'] = df['road'].map(value_mapping)
df.head()
```

```python
from sklearn.neighbors import KDTree
import numpy as np
```

```python
coordinates = df[['latitude', 'longitude']].values
y_true = df['road'].values
y_pred = df['y_pred'].values
print(y_pred)
```

```python
radius = 0.002
```

```python
kdtree = KDTree(coordinates)
```

```python
def assign_majority_label(index):
    # Query KD-tree to find neighbors within radius
    neighbors_indices = kdtree.query_radius([coordinates[index]], r=radius)[0]
    
    print(neighbors_indices)
    
    # Extract labels of neighbors
    neighbor_labels = y_pred[neighbors_indices]
    
    print(neighbor_labels)
    
    # Determine majority label
    majority_label = np.bincount(neighbor_labels).argmax()
    
    print(majority_label)
    
    return majority_label
```

```python
assigned_labels = [assign_majority_label(i) for i in range(len(coordinates))]

df['y_pred'] = assigned_labels
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Compute accuracy
accuracy = accuracy_score(df['road'], df['y_pred'])
print("Accuracy:", accuracy)
report = classification_report(df['road'], df['y_pred'], output_dict=True)

precision_weighted_avg = report['weighted avg']['precision']
print("Precision: ", precision_weighted_avg)

recall_weighted_avg = report['weighted avg']['recall']
print("Recall: ", recall_weighted_avg)

f1_score_weighted_avg = report['weighted avg']['f1-score']
print("F1-score: ", f1_score_weighted_avg)

# Compute confusion matrix
cm = confusion_matrix(df['road'], df['y_pred'])
print("Confusion Matrix:")
print(cm)
```
